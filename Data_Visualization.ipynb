```python
# Convert to date format
data.COMPLAINT_DATE = pd.to_datetime(data.COMPLAINT_DATE)
# Add YEAR column for later use
data["YEAR"] = data.COMPLAINT_DATE.dt.year
data["DAY"] = data.COMPLAINT_DATE.dt.day
data["DAY_OF_WEEK"] = data.COMPLAINT_DATE.dt.dayofweek
data.index = pd.DatetimeIndex(data["COMPLAINT_DATE"])
```

##  Distribution of complaints per MONTHS.

```python
sns.countplot(x="COMPLAINT_MONTH", data=data, palette=p)
```

```python
sns.countplot(x="COMPLAINT_HOUR", data=data, palette=p)
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
complaints_pt = data.pivot_table(values='YEAR', index='DAY', columns='COMPLAINT_MONTH', aggfunc=len)
complaints_pt_year_count = data.pivot_table(values='YEAR', index='DAY', columns='COMPLAINT_MONTH', aggfunc=lambda x: len(x.unique()))
complaints_avg = complaints_pt / complaints_pt_year_count
complaints_avg.columns = months
plt.figure(figsize=(10,12))
plt.title('Average Number of Complaints per Day and Month', fontsize=14)
sns.heatmap(complaints_avg.round(), cmap='magma', linecolor='grey',linewidths=0.1, cbar=True, annot=True, fmt=".0f");
```

* The worst day is Feb 29th (well, this only happens every four years)
* So 27 complaints on July 16 is a more reasonable estimate
* The day with the least complaints is Christmas Day, nobody wants to spoil the celebration!
